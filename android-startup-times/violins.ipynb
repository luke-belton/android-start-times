```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
times = pd.read_excel('./bugsnag-start-times.xlsx')
```

```python
f, ax = plt.subplots(figsize=(30, 10))
sns.set_theme(style="whitegrid")
sns.boxplot(x="version", y="start_time",
            data=times)
sns.despine(left=True, bottom=True)
ax.set_ylim(30, 200)
ax.set_xlabel('Bugsnag Version')
ax.set_ylabel('Bugsnag Start time')
```
